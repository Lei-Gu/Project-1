```python
# Dependencies

%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager as fm
import matplotlib
import scipy.stats as st
import seaborn as sns
from sklearn import datasets
import os

import warnings
warnings.filterwarnings('ignore')
```

```python
# Name of the CSV file
csv_path = 'googlestore_clean.csv'
```

```python
# The read the CSV file in pandas
googlestore = pd.read_csv(csv_path)
googlestore.head()
```

```python
googlestore_df= googlestore[googlestore.Price != 0]
googlestore_df.head()
```

```python
#What is the average price per category? (discounting free apps)
average_price = googlestore_df.groupby(["Category"])['Price'].mean()
average_price_sorted = average_price.sort_values(ascending = False)
average_price_df = pd.DataFrame({'Average Price': average_price_sorted})
average_price_df.style.format({'Average Price': '${:.2f}'})
```

```python
#What distribution of free v. paid apps across categories
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
#labels = googlestore_df['Category'].value_counts().index
#data = googlestore_df['Category'].value_counts()

types = googlestore['Type'].value_counts()
type_labels= googlestore['Type'].value_counts().index

matplotlib.rcParams['font.size'] = 15
fig, ax = plt.subplots(figsize=(8, 8))

#ax.pie(data,  autopct='%1.1f%%', startangle=90, pctdistance=0.75)
#plt.legend(labels, loc="center")

ax.pie(types,labels=type_labels, radius=0.8,startangle=90, autopct='%1.1f%%', labeldistance=1.05)

#draw circle
#centre_circle = plt.Circle((0,0),0.5,color='black', fc='white',linewidth=0)
#fig = plt.gcf()
#fig.gca().add_artist(centre_circle)
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.savefig(os.path.join('paid_free_percents.png'))
plt.show()
```

```python
csv_path2 = 'free_paid_cleaned.csv'
```

```python
free_paid = pd.read_csv(csv_path2)
free_paid
```

```python
fig= plt.figure(figsize=(10,6))
plt.plot( 'Category', 'Free', data=free_paid, marker='o', markerfacecolor='blue', markersize=5, 
         color="black", linewidth=1)

plt.plot( 'Category', 'Paid', data=free_paid, marker='o', markerfacecolor='orange', markersize=5, 
         color="black", linewidth=1)

plt.title('Distribution of Paid and Free Apps by Category', fontsize = 20)
plt.ylabel('Number of Apps')
plt.xticks(rotation='vertical', fontsize=10)
plt.legend(loc='best')
plt.savefig(os.path.join('paid_free_price.png'))
plt.show()
```
